```python
import numpy as np
```

```python
import pandas as pd
BicycleWeather = pd.read_csv('C:\\Python Class\\Datascience\\Datasets\\BicycleWeather.csv')
BicycleWeather.head()
```

```python
# Extraction of PRCP column from the pandas object.
Bicycle_weather = np.array(BicycleWeather['PRCP'])
print("PRCP \n",Bicycle_weather)
```

```python
#  using boolean functions/operators to separate numbers from NAN
Bicycle_weather
val = pd.notna(Bicycle_weather)
# Bicycle_weather[val]
val
```

```python
#  using boolean masking to separate all zeros and non-zeros into different variables
zero = Bicycle_weather[val]
value = (zero == 0)
value2 = (zero != 0)
# zero[value]
print("All zeros: ", (zero[value]))
print("Non zeros: ", (zero[value2]))
```

```python
# From the variable holding non-zeros, multiply every values in the array with 100 and divide by 250
(zero[value2]) * 100 / 250 
```
